```python
# GroupBy Operation
data1 = { 'Name':['Jai','Manan','Arjun','Ramial','Jayanti', 'Arjun','Manan'],
         'Age':['20','23','18','40','45','22','12'],
         'Address':['Surat','Ahmedbad','Vadodara','Kalol','Kutch','Ahmedbad','Ahmedbad']
}
```

```python
type(data1)
```

```python
# Convert dictanry data type into DataFrame

df_new = pd.DataFrame(data1)
df_new
```

```python
# GroupBy the data by Name
df_new.groupby('Name')
print(df_new.groupby('Name').groups)
```

```python
gk = df_new.groupby('Name')
gk
```

```python
list(gk)
```

```python
for i in gk:
    print(i)
```

```python
# GroupBy using multiple keys
df_new.groupby(['Name','Address'])
print(df_new.groupby(['Name','Address']).groups)
```

```python
gk1 = df_new.groupby(['Name','Address'])
gk1
```

```python
list(gk1)
```

```python
# Merging two data frames
temp_df = pd.DataFrame({ 'City':['Mumbai','Delhi','Banglore','Ahmedbad'],
                        'Temperature':['32','45','30','40'] })
temp_df
```

```python
humd_df = pd.DataFrame({ 'City':['Mumbai','Delhi','Banglore','Ahmedbad'],
                        'Humidity':['32','45','30','40'] })
humd_df
```

```python
# Merge 2 above DataFrame
tmp_humd = pd.merge(temp_df,humd_df, on='City')
tmp_humd
```
